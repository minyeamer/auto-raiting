# Hard-Voting

```python
import numpy as np
import pandas as pd
import random
import re
import os
from collections import Counter

SEED = 42
random.seed(SEED)
os.environ['PYTHONHASHSEED'] = str(SEED)
np.random.seed(SEED)
```

## Load History

```python
submission_dir = './data/samples/old/'
tree = list(os.walk(submission_dir))

def alias(name: str):
    name = name.replace('_with','').replace('_test','').replace('.csv','')
    name = ''.join([word[0].upper() for word in name.split('_')])
    return name

history = {alias(file):submission_dir+file for file in tree[0][2] if file not in {'.DS_Store'}}

koelectra_dir, checkpoints = tree[1][0]+'/', tree[1][2]
history.update({'KV'+re.findall('epoch(.*?).csv',file)[0]:koelectra_dir+file for file in checkpoints})

for name,path in history.items():
    history[name] = pd.read_csv(path)['target'].tolist()

history = pd.DataFrame(history)
history = history[sorted(history.columns)]
history.head()
```

## Submission

```python
submission = pd.read_csv('./data/sample_submission.csv')
voters = ['KVV','KVC','KVVE','KVCE','RL']
election = '+'.join(voters)

for idx in range(len(history[voters])):
    submission['target'][idx] = Counter(history[voters].iloc[idx]).most_common(1)[0][0]
submission.head()
```

```python
submission.to_csv(f'./data/samples/hard_voting_{election}.csv', index=False)
```
